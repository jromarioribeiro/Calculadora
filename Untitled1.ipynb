```python
# Definindo as funções para as operações matemáticas
def calcular(numero1, numero2, opcao):
    if opcao == '1':
        return numero1 + numero2
    elif opcao == '2':
        return numero1 - numero2
    elif opcao == '3':
        return numero1 * numero2
    elif opcao == '4':
        if numero2 != 0:
            return numero1 / numero2
        else:
            return "Erro: Divisão por zero!"
    else:
        return "Operação inválida!"

# Loop para o usuário realizar infinitas operações
while True:
    # transformando entradas em numeros reais para realizar as operações
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))

    # Exibindo todas as opções de operações
    print("Escolha a operação que deseja realizar:")
    print("1 - Adição")
    print("2 - Subtração")
    print("3 - Multiplicação")
    print("4 - Divisão")
    print("5 - Sair")
    opcao = input("Digite o número da operação: ")

    # opção para o usuário sair
    if opcao == '5':
        print("Saindo do programa.")
    # Para que o loop seja  interrompido
        break
    # Calculando o resultado
    resultado = calcular(num1, num2, opcao)

    # Exibindo o resultado
    print("O resultado é: " + str(resultado))

```
